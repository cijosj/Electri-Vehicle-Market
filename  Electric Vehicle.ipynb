```python
#read the csv file into a pandas dataframe
df = pd.read_csv(csv_location)
df
```

```python
# Creating a copy of dataframe as backup
df_copy = df
```

```python
# Renaming the column names for better readability
# Save the renamed column as a dataframe with new variable
data = df.rename(columns= {'VIN (1-10)': 'VIN', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'CAFV Eligibility','DOL Vehicle ID': 'DOL ID'})
data
```

```python
# Drop the Base MSRP column from the data as most of the values in the column is zero
EV_data = data.drop(['Base MSRP'],axis = 1)
EV_data
```

```python
# Convert the data type of postal code from float to int and also replacing NA values with zero
EV_data['Postal Code'] = EV_data['Postal Code'].fillna(0).astype(int)   
```

```python
# Print the data by 'Make','Electric Vehicle Type','Model','Model Year'
EV=EV_data[['Make','Electric Vehicle Type','Model','Model Year']]
EV
```

```python
# Count of EVs(BEV, PHEV) based on the type
EV_counts= EV['Electric Vehicle Type'].value_counts()
EV_counts
```

```python
# Top makes by model
# value_counts count the occurences of each combination of make and model
top_make_by_model= EV[['Make','Model']].value_counts()
top_make_by_model
```

```python
# Top 10 rows
top_make=top_make_by_model.head(10)
top_make
```

```python
#Create a bar-chart
# Looking for number of EVs by Make and Model
import matplotlib.pyplot as plt
# Kind of chart and its size
top_make.plot(kind='bar', figsize=(8,6))
# Add a label for the x axis
plt.xlabel("Make by model")
# Add a label for the y axis
plt.ylabel("Number of EVs")
# Add a title for your chart
plt.title("Top 10 EVs by make and model")
# Plot the chart
plt.show()
```

```python
# Print the data by 'Make','Electric Vehicle Type','State','City
EV=EV_data[['Make','Electric Vehicle Type','State','City','County']]
EV
```

```python
# EVs per location
EV_location = EV.groupby(['State','City','County'])[['Electric Vehicle Type']].count()
EV_location
```

```python
# Location(state/city/county) for EVs
# Sort the data in descending order
EV_location1= EV_location.sort_values(by=[('Electric Vehicle Type')], ascending=False)
EV_location1.reset_index()
```

```python
#Top 10 state,city, and county for EVs
location= EV_location1.head(10)
location
```

```python
#Create a bar-chart
#Looking for number of electric vehicles by location
import matplotlib.pyplot as plt
# Kind of chart and its size
location.plot(kind='bar', figsize=(8,6), color='maroon')
# Add a label for the x axis
plt.xlabel("Location of EVs")
# Add a label for the y axis
plt.ylabel("Number of EVs")
# Add a title for your chart
plt.title("Electric vehicles by location")
# Plot the chart
plt.show()
```

```python
# Electric range by make
# Print the data by 'Make','Electric Range','Model Year'
EV_range=EV_data[['Make','Electric Range','Model Year']]
EV_range
```

```python
# Range of EVs
EV_range1 = EV_range.groupby(['Make'])[['Electric Range']].count()
EV_range1
```

```python
#Sort values in descending order
speed = EV_range1.sort_values(['Electric Range'],ascending=False)
speed
```

```python
#Top 10 EV ranges
speed1= speed.head(10)
speed1
```

```python
#Create a bar-chart
import matplotlib.pyplot as plt
# Kind of chart and its size
speed1.plot(kind='bar', figsize=(8,6), color='orange')
# Add a label for the x axis
plt.xlabel("Make")
# Add a label for the y axis
plt.ylabel("Range of EVs")
# Add a title for your chart
plt.title("Electric vehicles range by make")
# Plot the chart
plt.show()
```
